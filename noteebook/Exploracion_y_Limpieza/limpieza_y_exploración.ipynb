```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Variables numéricas y categóricas

```python
# Obtener información sobre las variables categóricas
descripcion_categoricas = df.drop(columns='ID').describe(include=['object'])

# Almacenar la descripción de las variables categóricas en archivo .csv
descripcion_categoricas.to_csv('Outputs/CSV/descripcion_categoricas.csv')

# Visualizar la descripción de las variables categóricas
descripcion_categoricas
```

3. Medidas de tendencia central

```python
# Obtener métricas de tendencia central (media, mediana, mínimo y máximo)
metricas_tendencia_central = df.drop(columns='ID').describe().loc[['mean', '50%', 'min', 'max']]

# Obtener la moda
moda = df.mode().iloc[0]
metricas_tendencia_central.loc['mode'] = moda

# Almacenar las métricas de tendencia central en un archivo .csv
metricas_tendencia_central.to_csv('Outputs/CSV/metricas_tendencia_central.csv')

metricas_tendencia_central
```

4. Métricas de dispersión

```python
# Obtener las métricas de dispersión (desviación estándar, rango intercuartílico) para las variables numéricas
metricas_dispersión = df.drop(columns='ID').describe().loc[['std', '25%', '50%', '75%']]
metricas_dispersión.loc['IQR'] = df.drop(columns='ID').describe().loc['75%'] - df.drop(columns='ID').describe().loc['25%']

# Almacenar las métricas de dispersión en un archivo .csv
metricas_dispersión.to_csv('Outputs/CSV/metricas_dispersión.csv')

metricas_dispersión
```

### Preparación y limpieza de datos

1. Reconocimiento y tratamiento de atributos con valores únicos o distintos

```python
# Identificar cantidad de valores únicos
df.nunique()

# Guardar resultados en un archivo CSV
valores_unicos = df.nunique()
valores_unicos.to_csv('Outputs/CSV/valores_unicos.csv', header=True)

# Filtrar columnas con un solo valor único
unique_cols = [col for col in df.columns if df[col].nunique() <= 1]

# Filtrar columnas con todos los valores distintos
distinct_cols = [col for col in df.columns if df[col].nunique() == df.shape[0]]

print(unique_cols)
print(distinct_cols)
```

2. Reconocimiento y tratamiento de atributos con valores faltantes

```python
# Identificar las proporciones porcentuales de Null values para cada columna
null_proportions = df.isnull().mean() * 100

# Guardar resultados en un archivo CSV
null_proportions.to_csv('Outputs/CSV/null_proportions.csv', header=True)

# Si hay columnas con más del 80% de valores nulos, eliminarlas
print("Columnas con más del 80% de valores nulos:")
cols_to_drop = null_proportions[null_proportions > 80].index
print(cols_to_drop)
df.drop(columns=cols_to_drop, inplace=True)
```

```python
# Para las variables numéricas con menos del 15% de valores nulos, se imputa la mediana
cols_to_impute = null_proportions[(null_proportions < 15) & (null_proportions > 0)].index

print("Columnas numéricas con menos del 15% de valores nulos:")
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    if col in cols_to_impute and null_proportions[col] > 0:
        median = df[col].median()
        print(f'Mediana de {col}: {median}')
        df.fillna({col: median}, inplace=True)
```

```python
# Para las variables categóricas con menos del 15% de valores nulos, se imputa la moda
cols_to_impute = null_proportions[(null_proportions < 15) & (null_proportions > 0)].index
print("Columnas categóricas con menos del 15% de valores nulos:")

for col in df.select_dtypes(include=['object']).columns:
    if col in cols_to_impute and null_proportions[col] > 0:
        mode = df[col].mode()[0]
        print(f'Moda de {col}: {mode}')
        df.fillna({col: mode}, inplace=True)
```

3. Reconocimiento y tratamiento de registros atípicos

Observando los boxplot de las variables, se pueden evidenciar valores atípicos en las variables:
- Testosterone
- UrineFlow1
- UrineFlow2 (Ya se eliminó)
- nPregnancies
- AlcoholDay
- SexNumPartnLife
- SexNumPartYear

```python
# Graficar boxplot para las variables Testosterone, UrineFlow1, UrineFlow2, nPregnancies, AlcoholDay, SexNumPartnLife, SexNumPartYear  
atipic_vars = ['Testosterone', 'UrineFlow1', 'nPregnancies', 'AlcoholDay', 'SexNumPartnLife', 'SexNumPartYear']

# Crear subplot para las variables atípicas
fig, axs = plt.subplots(nrows=1, ncols=len(atipic_vars), figsize=(20, 5))
for i, var in enumerate(atipic_vars):
    sns.boxplot(x=df[var], ax=axs[i])
    axs[i].set_title(var)
plt.tight_layout()
plt.show()
```

```python
# Identificar y eliminar valores atípicos
def winsorizar_serie(serie, limite_inferior=0.01, limite_superior=0.99):
    q_inf = serie.quantile(limite_inferior)
    q_sup = serie.quantile(limite_superior)
    return np.clip(serie, q_inf, q_sup)

for var in atipic_vars:
    df[var] = winsorizar_serie(df[var])

# Crear subplot para las variables atípicas
fig, axs = plt.subplots(nrows=1, ncols=len(atipic_vars), figsize=(20, 5))
for i, var in enumerate(atipic_vars):
    sns.boxplot(x=df[var], ax=axs[i])
    axs[i].set_title(var)
plt.tight_layout()
plt.show()
```

4. Reconocimiento y tratamiento de atributos redundantes

```python
# Eliminar duplicados
df = df.drop_duplicates()

# Variables numéricas
numeric_cols = df.drop(columns=['ID']).select_dtypes(include=['float64', 'int64']).columns

# Calcular matriz de correlación
correlation_matrix = df[numeric_cols].corr()
correlation_matrix.to_csv('Outputs/CSV/correlation_matrix.csv', header=True)

# Identificar variables altamente correlacionadas
threshold = 0.90
highly_correlated = correlation_matrix[correlation_matrix.abs() >= threshold].stack().index.tolist()

# De la lista eliminar las tuplas que tienen la misma variable
highly_correlated = [(x, y) for x, y in highly_correlated if x != y]

highly_correlated
```
